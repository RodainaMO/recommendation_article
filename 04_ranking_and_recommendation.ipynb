```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load similarity scores file
df = pd.read_csv('../results/similarity_scores.csv')

# Sort and get top N recommendations
top_n = 10
recommended = df.sort_values(by='similarity', ascending=False).head(top_n)

# Save recommendations
recommended.to_csv('../results/sample_recommendations.csv', index=False)

# Optionally save feedback notes
with open('../results/user_feedback_notes.txt', 'w') as f:
    f.write("User liked the recommendations based on keywords: ...")  # placeholder

```

```python


# Plot distribution of similarity scores
plt.figure(figsize=(10, 6))
sns.histplot(df['similarity'], kde=True, bins=30, color='blue')
plt.title('Distribution of Similarity Scores')
plt.xlabel('Cosine Similarity')
plt.ylabel('Frequency')
plt.savefig('../results/similarity_distribution.png')
plt.show()




```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=recommended, x='Title', y='similarity', hue='Category', dodge=False, palette='viridis')
plt.xticks(rotation=90)
plt.title(f'Top {top_n} Recommended Articles by Category')
plt.xlabel('Article Title')
plt.ylabel('Cosine Similarity')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('../results/top10_articles_by_category.png')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='similarity', hue='Category', palette="Set2", legend=False)
plt.title('Distribution of Similarity Scores by Category')
plt.xlabel('Category')
plt.ylabel('Cosine Similarity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../results/similarityxcategory.png')
plt.show()

```
